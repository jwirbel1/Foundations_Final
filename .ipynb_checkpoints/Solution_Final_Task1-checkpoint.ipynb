Previously and also as part of forthcoming courses, we will use a popular library for Machine learning called NumPy. The NumPy library allows a user to specify an array, create arrays of values 0s and 1s, provide implementations for the dot product, matrix multiplication, and much more. As part of this question, we will create our own implementation of a few functionalities supported by the NumPy Library. We will call our implementation SNumPy (for Sub-NumPy). SNumPy will be the name of the class you implement, and we will refer to it by the shorthand “snp” from here on. You are asked to provide an implementation in Python for the following few selected functionalities:

1 snp.ones(Int): the ones function takes an int parameter and returns an array (list) of length int parameter, and the array contains only ones. Example: snp.ones(5) = [1,1,1,1,1]

2 snp.zeros(Int): similar to the ones function, expect returns an array of zeros instead of ones.

3 snp.reshape(array, (row, column)): takes an array and converts it into the dimensions specified by the tuple (row, column). Hence this function converts from a vector to a matrix. For an example on reshape functionality of numpy, refer to fig. 1.

4 snp.shape(array) : returns a tuple with the matrix/vector’s dimension e.g. (# rows, # columns)

5 snp.append(array1, array2) : returns a new vector/matrix that is the combination of the two input vectors/matrices. Note that you can’t append a vector to a matrix and vice versa and therefore use suitable exception handling and throw/return user friendly error messages.

6 snp.get(array,(row,column)):returnsthevaluespecifiedbythecoordinatepoint(row,column) of the array provided (can be vector or matrix).

7 snp.add(array1, array1) : addition on vectors/matrices

8 snp.subtract(array1, array1) : subtraction on vectors/matrice

9 snp.dotproduct(array1, array1): computes the dot product between two arrays (which could be vectors or/and matrices) and returns an appropriate value. Use appropriate exception handling to output user-friendly error messages if the dot product cannot be performed between the given arrays.

10 Optional: Gauß

```python
class snp():
    """
    This class contains all functions for SNumPy.
    """

    def ones(number: int):
        """
        This method takes an integer and returns a list of ones of lenght of that integer.
        Arguments:
            number:length of the resulting list
        Returns: 
            list: list with 0s of the given length
        """
        
        try:
            assert isinstance(number, int), "Input must be a positive integer." # Check that the input is valid
            if number <= 0:
                return "Integer must be positive."
            ones = [1 for x in range(number)] # create list that contain input number times 1
            return ones # return list
        
        except AssertionError as msg:
            print(msg)
    
    def zeros(number: int):
        """
        This method takes an integer and returns a list of ones of lenght of that integer.
        Arguments:
            number:length of the resulting list.
        Returns: 
            list: list with 1s of the given length
        """
        try:
            assert isinstance(number, int), "Input must be a positive integer." # Check that the input is valid
            if number <= 0:
                return "Integer must be positive."
            zeros = [0 for x in range(number)] # create list that contain input number times 1
            return zeros # return list
        
        except AssertionError as msg:
            print(msg)
    
    
    def reshape(array: list, shape: tuple, order = "row"):      	
        """
        This method takes an array and converts it into the dimensions specified by the tuple (row, column).
        Arguments:
            array: array that has to be converted
            shape: tuple that includes new shape (row, column)
        Returns: 
            array_final: reshaped array with the new dimensions
        """
        
        try:
            assert isinstance(shape, tuple), "Shape must be a tuple"  # Check that the input is valid
            assert isinstance(array, list), "Array must be a list"    # Check that the input is valid
            assert (order == "row" or order == "column"), "The order has to be row or column" # Check for correct axis
            try:
                array[0][0]
                array = sum(array,[])
            finally:
                assert (shape[0]*shape[1]) == len(array), "The dimensions of the array do not fit the transforming tuple" # Check for correct dimensionality
                rows = shape[0]           # number of rows is the first input in the shape tuple
                columns = shape[1]        # number of columns is the second input in the shape tuple
                array_final = []          # create final array / list
                
                if order == "row":
                    for i in range(rows):                       # for rows times
                        row = []                               
                        for j in range(columns):                # for columns times
                            row.append(array[columns*i + j])    # append the list with the elements of the input array (columns times the current row number plus the column number -> e.g. first time: array[3*0 + 1 -> 1, +2 -> 2 ...])
                        array_final.append(row)                 # append the new row list to final array list
                    return array_final                      
                
                if order == "column":
                    for i in range(rows):          # iterate over rows
                        row = []                   
                        for j in range(columns):     # Iterate over elements in row[columnindex]
                            row.append(array[(j * rows) + i]) # append element to new index in in column position
                        array_final.append(row)      # append new column to final output
                    return array_final
        
        except AssertionError as msg:
            print(msg)


    def shape(array: list):
        """
        This method takes an array and returns a tuple with the matrix/vector’s dimension (# rows, # columns)
        Arguments:
            array: list that dimensions need to be discovered
        Returns:
            (row, column[0]): tuple with the shape of the array
        """
        
        # Check that the input is a list
        if not isinstance(array, list):
            raise TypeError("Array must be a list!")

        # Check if the array is a 1D or 2D array
        try:
            array[0][0]  # Check if the array is a 2D array
            is_2d = True
        except:
            is_2d = False

        # Check that the array is a 2D array with equal length for all elements (if it is a 2D array)
        if is_2d:
            if any(len(array[0]) != len(i) for i in array):
                raise ValueError("Error: Elements in the list have not the same length")

        # Return the dimensions of the array
        if is_2d:
            return (len(array), len(array[0]))
        elif len(array) > 1:
            return (1, len(array))
        else:
            return () # empty tuple, if array was empty
          

    def get(array: list, place: tuple):
        """
        This method accesses an array element by referring to its index number.
        Input:
            array1: the first array or vector
            place: the index of the requested element
        Returns:
            One element
        """
        
        # Check that the input is valid
        if not isinstance(array, list):
            raise TypeError("Array must be a list!")
        if not isinstance(place, tuple):
            raise TypeError("Place must be a tuple!")

        # Check if the array is a 1D or 2D array
        try:
            array[0][0]  # Check if the array is a 2D array
            is_2d = True
        except:
            is_2d = False

        # Check that the array is a 2D array with equal length for all elements (if it is a 2D array)
        if is_2d:
            if any(len(array[0]) != len(i) for i in array):  # Check compare the length of each element in the 2d array with the first
                raise ValueError("Error: Elements in the list have not the same length")

        # Check that the specified position is within the bounds of the array
        if is_2d:
            if place[0] >= len(array) or place[1] >= len(array[0]):
                raise TypeError("Out of bound")
        else:
            if place[0] >= len(array):
                raise TypeError("Out of bound")

        # Return the value at the specified position in the array
        if is_2d:
            return array[place[0]][place[1]]
        else:
            return array[place[0]]

    def check_integer_or_float(array: list):
        """
        This method checks if there are elements in the array which are not integers
        Arguments:
            array: list (1d), or list of list (2d)
        Returns:
            True:  If all elements in the list or list of list are integers
            False: If there is at least one element which is not an integer
        """
        
        # Check that the input is a list
        if not isinstance(array, list):
            raise TypeError("Input must be a list!")

        # Check whether the elements in the array are integers
        if all(isinstance(x, list) for x in array):
            # The input is a 2D array
            return all(isinstance(i,(int, float)) for row in array for i in row)
        else:
            # The input is a 1D array
            return all(isinstance(i, (int, float)) for i in array)
   
    def add(array1: list, array2: list):
        """
        This method adds two arrays.
        Input:
            array1: the first array or vector
            array2: the second array or vector
        Returns:
            Array with added elements
        """

        # Check that the input is a list of integers
        if not snp.check_integer_or_float(array1) or not snp.check_integer_or_float(array2):
            raise ValueError("Input arrays must contain integers or floats!")

        array1_shape = snp.shape(array1)
        array2_shape = snp.shape(array2)

        try:
            assert array1_shape == array2_shape, "Dimensions of both inputs are not equal"
            if array1_shape[0] == 1:
                return [sum(x) for x in zip(array1,array2)]
            
            else:
                array_added = []
                for i in range(len(array1)):
                    array_sub = []
                    for x in range(len(array1[i])):
                        array_sub.append(array1[i][x] + array2[i][x])
                    array_added.append(array_sub)
                
                return array_added
            
        except AssertionError as msg:
            print(msg) 
    
    def subtract(array1: list, array2: list):
        """
        This method subtracts two arrays.
        Input:
            array1: the first array or vector
            array2: the second array or vector 
        Returns:
            Array with subtracted elements
        """

        # Check that the input is a list of integers
        if not snp.check_integer_or_float(array1) or not snp.check_integer_or_float(array2):
            raise ValueError("Input arrays must contain integers or floats!")

        array1_shape = snp.shape(array1)
        array2_shape = snp.shape(array2)

        try:
            assert array1_shape == array2_shape, "Dimensions of both inputs are not equal"
            if array1_shape[0] == 1:
                return [x - i for x,i in zip(array1,array2)]
            
            else:
                array_added = []
                for i in range(len(array1)):
                    array_sub = []
                    for x in range(len(array1[i])):
                        array_sub.append(array1[i][x] - array2[i][x])
                    array_added.append(array_sub)
                return array_added
            
        except AssertionError as msg:
            print(msg) 

    def dotproduct(array1: list, array2: list):
        """
        This method returns the dot product of two vectors, matrices or a vector and a matrix.
        Input:
            array1: the first array or vector
            array2: the second array or vector
        Returns:
            dot_product
        """
        
        # Check that the input is valid
        try:
            assert snp.check_integer_or_float(array1) and snp.check_integer_or_float(array2), "Inputs must be numberical"
            # Vector operations
            if isinstance(array1[0], int) and isinstance(array2[0], int):
                try: 
                    assert len(array1) == len(array2), "Vectors must have the same shape!"
                    value_beg = 0
                    # Get elementwise values from both arrays
                    array1 = [int(x) for x in array1]
                    array2 = [int(x) for x in array2]
                    # Perform bitwise multiplication and add it to final value
                    for element in range(len(array1)):
                        value_beg += array1[element] * array2[element]
                
                    return value_beg  
            
                except AssertionError as msg:
                    print(msg)
            
            # Matrix / vector operation, first input a vector
            elif (isinstance(array1[0], int) and isinstance(array2[0], list)):
                try:
                    assert len(array1) == len(array2), "Array and vector must be compatible"
                    array_dot = [] # initialize return vector
                    for column_2 in range(len(array2[0])):
                        value_beg = 0 # initialitze element for each position of return vector
                        for column_1 in range(len(array1)):
                            # Perform bitwise multiplication of the 
                            value = array1[column_1]*array2[column_1][column_2]
                            value_beg += value
                        array_dot.append(value_beg)
                    return array_dot
                    
                except AssertionError as msg:
                    print(msg)
            
            # Matrix / vector operation, second input a vector
            elif (isinstance(array1[0], list) and isinstance(array2[0], int)):
                try:
                    assert len(array1) == len(array2), "Array and vector must be compatible"
                    array_dot = [] # initialize return vector
                    for column_2 in range(len(array1[0])):
                        value_beg = 0 # initialitze element for each position of return vector
                        for column_1 in range(len(array2)):
                            # Perform bitwise multiplication of the 
                            value = array2[column_1]*array1[column_1][column_2]
                            value_beg += value
                        array_dot.append(value_beg)
                    return array_dot
                    
                except AssertionError as msg:
                    print(msg)
                
            # Matrix / Matrix operation        
            elif isinstance(array1[0], list) and isinstance(array2[0], list):
                try:
                    print(snp.shape(array1), snp.shape(array2))
                    assert snp.shape(array1)[1] == snp.shape(array2)[0], "Matrices must have compatible shapes (#cols array 1 = #rows array2)!"
                    array_dot = [] # Initialize return vector
                    for row_1 in range(len(array1)):
                        array_dot_sub = []
                        for column_2 in range(len(array2[1])):
                            value_beg = 0
                            for column_1 in range(len(array1[0])):
                                # Iterate through the rows/cols for array1 and cols/rows for array2
                                #print(column_1, column_2, row_1)
                                value = array1[row_1][column_1]*array2[column_1][column_2]
                                value_beg += value
                            array_dot_sub.append(value_beg)
                        array_dot.append(array_dot_sub)
                
                    return array_dot

                except AssertionError as msg:
                    print(msg) 

        except AssertionError as msg:
            print(msg)    
    
    def append(array1: list, array2: list, axis = None):
        """
        This function appends the second input to the first, given they are compatible
        Input:
            array1: first input, can be a Matrix or a Vector
            array2: second input, can be a Matrix or a Vector
            axis: This defines the axis, on which the second input will be appended to the first
        Returns:
            output: input2 appended to input1 at the specified axis
        """
        
        # Check that the input is valid
        try:
            assert isinstance(array1, list) and isinstance(array2, list), "Inputs must be arrays" 
            # If the second input is to be appended "behind" the first one:
            if axis == 0:                           
                # in case both are Matrices
                if isinstance(array1[0], list) and isinstance(array2[0], list):
                    try:
                        assert snp.shape(array1)[1] == snp.shape(array2)[1], "Matrices must have the same dimensionality"
                        array_append = [] # initialize return array
                        for vector in array1: # append all contents of first input to this new array
                            array_append.append(vector)
                        for vector in array2: # append all contents of the second input to this new array
                            array_append.append(vector)
                        return array_append
                        
                    except AssertionError as msg:
                        print(msg)
                # Both are Vectors
                elif (isinstance(array1[0], int) and isinstance(array2[0], int)) or (isinstance(array1[0], str) and isinstance(array2[0], str)):
                    for element in array2: # append all elements of input 2 to the end of input 1
                        array1.append(element)
                            
                    return array1

            elif axis == 1:
                    
                # Both are Matrices:
                if isinstance(array1[0], list) and isinstance(array2[0], list):
                    try: 
                        assert snp.shape(array1) == snp.shape(array2), "Both input matrices must have the same dimensionality"

                        for i in range(len(array1)):
                            for x in array2[i]:
                                array1[i].append(x)
                        return array1
                            
                    except AssertionError as msg:
                        print(msg)
                            
                # Both are Vectors
                elif (isinstance(array1[0], int) and isinstance(array2[0], int)) or (isinstance(array1[0], str) and isinstance(array2[0], str)):
                    for i in array2:
                        array1.append(i)
                    return array1
                                    
                else:
                    print(ValueError("Inputs can't be vector and matrix"))
                        
            else:
                # No axis is given
                # Both are Matrices:
                if isinstance(array1[0], list) and isinstance(array2[0], list):
                    return_array = [] # initialize return array
                    for i in range(len(array1)): # add all elements from the first array
                        for x in array1[i]:
                            if isinstance(x, list):
                                for element in x:
                                    return_array.append(element)
                            else:
                                return_array.append(x)
                    for i in range(len(array2)): # add all elements from the second array
                        for x in array2[i]:
                            if isinstance(x, list):
                                for element in x:
                                    return_array.append(element)
                            else:
                                return_array.append(x)   
                    return return_array
                            
                # Both are Vectors
                elif (isinstance(array1[0], int) and isinstance(array2[0], int)) or (isinstance(array1[0], str) and isinstance(array2[0], str)):
                    for i in array2: # add all elements of the second input to the first
                        array1.append(i)
                    return array1
                                
                # Matrix and vector
                elif (isinstance(array1[0], list) and isinstance(array2[0], int)) or (isinstance(array1[0], list) and isinstance(array2[0], str)):
                    output = []
                    for vector in array1:
                        for i in vector:
                            output.append(i)
                    for element in array2:
                        output.append(element)
                    
                    return output
        except AssertionError as msg:
                        print(msg)     

    def linalg(array1: list, array2: list):
        """
        This function solves a system of linear equasions utilizing the Gauss algorithm and row reduction.
        Inputs:
            array1: the coefficient matrix (a)
            array2: the ordinate matrix (x)
        Returns:
            result: Solution to the system ax = b, Returned shape is identical to b
        """
          
        # Check that the input is valid
        try:
            assert isinstance(array1[0], list), "Input 1 must be an array"
            assert isinstance(array2[0], list), "Input 2 must be an array"
            assert snp.shape(array1)[0] == snp.shape(array2)[1], "Input 1 and 2 must be compatible"
            assert snp.check_integer_or_float(array1) and snp.check_integer_or_float(array2), "Inputs must be numerical and not nested"

            # build the combined matrix
            array3 = []
            for index, vector in enumerate(array1):
                vector.append(array2[0][index])
                array3.append(vector)

            n = snp.shape(array3)[1] # Get the number of cols
            n2 = snp.shape(array3)[0] # Get the number of rows
            index = list(range(n2)) # get the indexes   

            for pivot in range(n2): # Iterate over the rows
                pscale = 1.0 / array3[pivot][pivot] # get the necessary scaler to bring the pivot element to 1

                for j in range(n): # adjust all elements in the pivot row with the scaler
                    array3[pivot][j] *= pscale
                            
                for i in index[0:pivot] + index[pivot+1:]: #iterate over the rest of the matrix, except for the pivot row
                    rscale = array3[i][pivot] # get the scaler for each row element to bring it to zero
                    for j in range(n):
                        array3[i][j] = array3[i][j] - rscale * array3[pivot][j] # adjust the row elements
            result = []
            for vector in array3:
                result.append(vector[-1])
                
            return result
                
        except AssertionError as msg:
                print(msg) 
```
