```python
def ones(number: int):
        """
        This method takes an integer and returns a list of ones of lenght of that integer.
        Arguments:
            number:length of the resulting list.
        Returns:
            self.ones: list of ones with the length specified by the input
        """
        try:
            assert isinstance(number, int), "Input must be an integer" # Input must be an integer!
            ones = [1 for x in range(number)] # create list that contain input number times 1
            return ones # return list
        
        except AssertionError as msg:
            print(msg)
```

```python
def zeros(number: int):
        """
        This method takes an integer and returns a list of zeros of lenght of that integer.
        Arguments:
            number: length of the resulting list.
        Returns:
            self.zeros: listof zeros with the length specified by the input
        """
        try:
            assert isinstance(number, int), "Input must be an integer" # Input must be an integer!
            zeros = [0 for x in range(number)] # create list that contain input number times 1
            return zeros # return list
        
        except AssertionError as msg:
            print(msg)
```

```python
def reshape(array: list, shape: tuple, order = "row"):

    """
    This method takes an array and converts it into the dimensions specified by the tuple (row, column).
    Arguments:
        array: array that has to be converted
        shape: tuple that includes new shape (row, column)
    Returns: 
        array_final: reshaped array with the new dimensions
    """
    
    try:
        assert isinstance(shape, tuple), "Shape must be a tuple" # shape must be a tuple!
        assert isinstance(array, list), "Array must be a list" # shape must be a list!            
        assert (shape[0]*shape[1]) != len(array), "The dimensions of the array do not fit the transforming tuple" # Check for correct dimensionality
        assert (order == "row" or order == "column"), "The order has to be row or column" # Check for correct axis
              
        rows = shape[0] # number of rows is the first input in the shape tuple
        columns = shape[1] # number of columns is the second input in the shape tuple
        array_final = [] #create final array / list
            
        if order == "row":
            for i in range(rows): #for rows times
                row = [] # create a list
                for x in range(columns): # for columns times
                    row.append(array[columns*i + x]) # append the list with the elements of the input array (columns times the current row number plus the column number -> e.g. first time: array[3*0 + 1 -> 1, +2 -> 2 ...])
                array_final.append(row) # append the new row list to final array list
            return array_final # return list
            
        if order == "column":
            for i in range(rows): # iterate over rows
                col = []
                for j in range(columns): # Iterate over elements in row[columnindex]
                  #index = (j * rows) + 1
                  col.append(array[(j * rows) + 1]) # append element to new index in in column position
                array_final.append(col) # append new column to final output
            return array_final
              
    except AssertionError as msg:
        print(msg)
```

```python
def dotproduct(array1: list, array2: list): # JULIUS
    """
    This method returns the dot product of two vectors, matrices or a vector and a matrix.
    Input:
        array1: the first array or vector
        array2: the second array or vector
    Returns:
        dot_product
    """
    
    # Vector operations
    if isinstance(array1[0], int) and isinstance(array2[0], int):
        try: 
            assert len(array1) == len(array2), "Vectors must have the same shape!"
            value_beg = 0
            # Get elementwise values from both arrays
            array1 = [int(x) for x in array1]
            array2 = [int(x) for x in array2]
            # Perform bitwise multiplication and add it to final value
            for element in range(len(array1)):
                value_beg += array1[element] * array2[element]
            
            return value_beg  
        
        except AssertionError as msg:
            print(msg)
        
    # Matrix / vector operation, first input a vector
    elif (isinstance(array1[0], int) and isinstance(array2[0], list)):
        try:
            assert len(array1) == len(array2), "Array and vector must be compatible"
            array_dot = [] # initialize return vector
            for column_2 in range(len(array2[0])):
                value_beg = 0 # initialitze element for each position of return vector
                for column_1 in range(len(array1)):
                    # Perform bitwise multiplication of the 
                    value = array1[column_1]*array2[column_1][column_2]
                    value_beg += value
                array_dot.append(value_beg)
            return array_dot
            
        except AssertionError as msg:
            print(msg)
    
    # Matrix / vector operation, second input a vector
    elif (isinstance(array1[0], list) and isinstance(array2[0], int)):
        try:
            assert len(array1) == len(array2), "Array and vector must be compatible"
            array_dot = [] # initialize return vector
            for column_2 in range(len(array1[0])):
                value_beg = 0 # initialitze element for each position of return vector
                for column_1 in range(len(array2)):
                    # Perform bitwise multiplication of the 
                    value = array2[column_1]*array1[column_1][column_2]
                    value_beg += value
                array_dot.append(value_beg)
            return array_dot
            
        except AssertionError as msg:
            print(msg)
        
    # Matrix / Matrix operation        
    elif isinstance(array1[0], list) and isinstance(array2[0], list):
        try:
            assert snp.shape(array1) == snp.shape(array2), "Matrices must have the same shape!"
            array_dot = [] # Initialize return vector
            for row_1 in range(len(array1)):
                array_dot_sub = []
                for column_2 in range(len(array2[0])):
                    value_beg = 0
                    for column_1 in range(len(array1[0])):
                        # Iterate through the rows/cols for array1 and cols/rows for array2
                        value = array1[row_1][column_1]*array2[column_1][column_2]
                        value_beg += value
                    array_dot_sub.append(value_beg)
                array_dot.append(array_dot_sub)
        
            return array_dot

        except AssertionError as msg:
            print(msg)        
```

```python
def append(array1, array2, axis = None):
    """
    This function appends the second input to the first, given they are compatible
    Input:
        array1: first input, can be a Matrix or a Vector
        array2: second input, can be a Matrix or a Vector
        axis: This defines the axis, on which the second input will be appended to the first
    Returns:
        output: input2 appended to input1 at the specified axis
    """
    try:
        assert isinstance(array1, list) and isinstance(array2, list), "Inputs must be arrays" 
        if axis == 0:
            if (isinstance(array1[0], list) and isinstance(array2[0], list)) and (len(array1) > 1 and len(array2) > 1):
                print(snp.shape(array1), snp.shape(array2))
                assert snp.shape(array1)[1] == snp.shape(array2)[1], "Inputs must have the same shape!"
                #for i in range(len(array1)):
                #    if len(array1[i]) != len(array1[0]):
                #        raise ValueError("The Matrices need to have vectors of the same length!")
                #    if len(array2[i]) != len(array1[0]):
                #        raise ValueError("The Matrices need to have vectors of the same length!")
                
                for vector in array2:
                    array1.append(vector)
                return array1
                              
            elif isinstance(array1[0], list) and isinstance(array2[0], list):
                    try:
                        assert snp.shape(array1)[1] == snp.shape(array2)[1], "Matrices must have the same dimensionality"
                        array_append = [array1[0], array2[0]]
                        return array_append
                    
                    except AssertionError as msg:
                        print(msg)

            # Both are Vectors
            elif isinstance(array1[0], int) and isinstance(array2[0], int):
                for element in array2:
                    array1.append(element)
                     
                return array1

        elif axis == 1:
            
            # Both are Matrices:
            if isinstance(array1[0], list) and isinstance(array2[0], list):
                try: 
                    assert len(array1) == len(array2), "Both input matrices have to be of the same length"
                    for i in range(len(array1)):
                        for x in array2[i]:
                            array1[i].append(x)
                    return array1
                        
                except AssertionError as msg:
                    print(msg)
                    
                # Both are Vectors
            elif isinstance(array1[0], int) and isinstance(array2[0], int):
                for i in array2:
                    array1.append(i)
                return array1
                            
            else:
                raise ValueError("Inputs can't be vector and matrix")
                    

        else:
            # No axis is given
            # Both are Matrices:
            if isinstance(array1[0], list) and isinstance(array2[0], list):
                l = []
                for i in range(len(array1)):
                    for x in array1[i]:
                        l.append(x)
                for i in range(len(array2)):
                    for x in array2[i]:
                        l.append(x)     
                return l  
                        
            # Both are Vectors
            elif isinstance(array1[0], int) and isinstance(array2[0], int):
                for i in array2:
                    array1.append(i)
                return array1
                        
            else:
                raise ValueError("Inputs can't be vector and matrix")
            
    except AssertionError as msg:
                print(msg)        
```

```python
def linalg(array1, array2): # get solution of linear equation
        '''
        Designed after functionality of numpy.linalg.solve:
        
        Input:
        array1: Coefficient matrix, must be a square and full-rank matrix
        array2: Ordinate or “dependent variable” values
        
        LinAlgError:
        If array1 is singular or not square
        '''
        try:
            #assert (snp.shape(array1)[0] == snp.shape(array1)[1]), "LinAlgError: Dimension of first input must be equal for rows and columns equal"
            assert (snp.shape(array2)[0] == 1), "Dimension of second input must be one dimensional"
            assert (snp.shape(array1)[1] == snp.shape(array2)[1]), "Dimension of second input must fit to first array"
                
            matrix = []
            for i in range(len(array1)): # get augmented matrix, combine both input (horizontal)
                matrix.append(array1[i])
                matrix[i].append(array2[i])

            matrix_shape = snp.shape(matrix) # get shape of the matrix 

            matrix.sort(reverse=True) # switch rows so that biggest values are on the top

            # to reduced echelon form
            for i in range(min(matrix_shape)): # goal: iterating through each diagonal of the matrix
                if matrix[i][i]!=0: # check if diagonal value is zero
                    diagonal = matrix[i][i]
                    for j in range(matrix_shape[1]): # when not zero, divide the diagonal by itself and all other values in the row
                        matrix[i][j] /= diagonal
                    for k in range(matrix_shape[0]): # operation so that the column will be a pivot column
                        if (matrix[k][i]!=0) and (k!=i): # skip the pivotized row
                            factor = matrix[k][i] 
                            for l in range(matrix_shape[1]):
                                matrix[k][l] -= factor * matrix[i][l] # substract with the above row so that pivot column value is 0

            # case when there are infinite solutions
            for i in range(matrix_shape[0]):
                if matrix[i][-1]!=0:
                    if matrix[i].count(0)==(matrix_shape[1]-1): # count if there are only zero values in a row
                        return "matrix has infinite solution"

            # case when there is no solution
            for i in range(matrix_shape[0]):
                if matrix[i][-1]==0:
                    if matrix[i].count(0)==(matrix_shape[1]-1): # if number of zeros equals the number of unknowns, then no solution
                        return "matrix has no solution"

            # case when there is only one solution
            array = []
            for i in range(matrix_shape[0]):
                array.append(matrix[i][-1])
            return "array(" + str(array) + ")"
        
        except AssertionError as msg:
            print(msg) 
```
